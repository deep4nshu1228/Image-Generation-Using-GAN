## Importing the Libraries

```python
import tensorflow as tf
import os
```

## Loading the Pre-trained Model 

```python
generator_model = tf.keras.models.load_model(os.path.join('/content/drive/MyDrive/GAN/Saved Model','generatormodel 100 epoch.h5'))
generator_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## Generating Image

```python
imgs = generator_model.predict(tf.random.normal((16, 128, 1)))
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(ncols=4, nrows=4, figsize=(10,10))
for r in range(4): 
    for c in range(4): 
        ax[r][c].imshow(imgs[(r+1)*(c+1)-1])
```
